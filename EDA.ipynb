```python
#Imports
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
#Import Data and Normalize

X = pd.read_csv("X.csv")
scaler = StandardScaler()
scaled_data = scaler.fit_transform(X)
X = pd.DataFrame(scaled_data, columns=X.columns)

y = pd.read_csv("y.csv")

y_classes = (y > 0).astype(int) #1 if improved, 0 otherwise
```

```python
#Regression EDA: Feature-Feature Correlation Heatmap
plt.figure(figsize=(12, 10))
cor = X.corr()
sns.heatmap(cor, annot=False, cmap=plt.cm.Reds)
plt.show()
```

```python
#Regression EDA: Feature-Output Scatters
fig, axes = plt.subplots(11, 4, figsize=(20, 40))
axes = axes.flatten()

for i, col in enumerate(X.columns):
    axes[i].scatter(X[col], y)
    axes[i].set_title(f'{col} vs. y')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('y')

plt.tight_layout()
plt.show()
```

```python
#Classification EDA: Feature-Output Scatters
fig, axes = plt.subplots(11, 4, figsize=(20, 40))
axes = axes.flatten()

for i, col in enumerate(X.columns):
    axes[i].scatter(X[col], y_classes)
    axes[i].set_title(f'{col} vs. y')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('y')

plt.tight_layout()
plt.show()
```
